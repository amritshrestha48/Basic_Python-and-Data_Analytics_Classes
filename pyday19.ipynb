# Bar Plot in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1)
plt.show()
```

```python
df1
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex")
plt.show()
```

# Order

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1)
plt.show()
```

# hue_order

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, hue_order = ['Female','Male'])
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1,order = order1, color = "green")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, palette = "deep")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, palette = "Accent")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, palette = "gist_earth")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1, palette = "prism")
plt.show()
```

# Saturation

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1,
            palette = "prism", saturation = 0.9)
plt.show()
```

# Histogram in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1
```

```python
sns.histplot(df1["flipper_length_mm"])  #also use sns.displot()
plt.show()
```

# Bins parameter

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240])
plt.show()
```

# Color

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], color = "red")
plt.show()
```

# kde

```python
sns.histplot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True,
            color = "red")
plt.show()
```
