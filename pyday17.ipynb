```python
#importing Libraries

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'Salman.jpg'

#opening image using piL

image = Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image, cmap = 'gray')
plt.title("Salman Khan")
plt.xlabel("Length")
plt.ylabel("Breadth")
plt.grid()
plt.show()
```

# Matplotlib- Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['English','Nepali','Maths','Science']

plt.pie(x,labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ['English','Nepali','Maths','Science']
c = ['yellow','blue','green','orange']
plt.pie(x, labels = y, colors = c)
plt.show()
```

```python
x = [10,20,30,40]
y = ['English','Nepali','Maths','Science']

plt.pie(x,labels = y)
plt.legend()
plt.show()
```
