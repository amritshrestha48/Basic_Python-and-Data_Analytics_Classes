# Seaborn Library

# Part-1

Seaborn is a library for making statistical graph in python.

It builds on top of matplotlib and integrates closely with pandas data structures.

plots are basically used for visualizing the relationship between variables.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
```

# matplotlib

```python
plt.plot(var,var1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
```

# Seaborn

```python
sns.lineplot(x = var, y = var1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]

import pandas as pd

df = pd.DataFrame({"var":var, "var1":var1})
sns.lineplot(x = 'var', y = 'var1', data=df)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex" )
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", palette = "rocket_r" )
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", style = "sex", 
             palette = "rocket_r", markers = ["o",">"])
plt.show()
```
