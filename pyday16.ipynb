# Matplotlib

# Part-1

Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.

The greatest benefits of visualization is that its allows us visual access to huge amounts of data in easily digestible visuals.

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'

plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y,marker = 'o')
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(y,marker = '>')
plt.plot(x,marker = 'o')
plt.show()
```

# Bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y, color = c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y, color = 'red')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y, color = "#4CAF50")   #COLOR CODE
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.bar(x,y, color = "#4CAF50", width = 0.2)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])

plt.barh(x,y, color = "#4CAF50", height = 0.1)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]
plt.title("scatter plot")
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]
plt.title("scatter plot")
plt.xlabel("months")
plt.ylabel("numbers")
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]
plt.title("scatter plot")
plt.xlabel("months")
plt.ylabel("numbers")
plt.scatter(x,y, color = "red")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]
plt.title("scatter plot")
plt.xlabel("months")
plt.ylabel("numbers")
c = ['red','blue','orange','green','yellow','brown']
plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,6]
plt.title("scatter plot")
plt.xlabel("months")
plt.ylabel("numbers")
c = ['red','blue','orange','green','yellow','brown']
plt.scatter(x,y, color = c, s = 300)
plt.show()
```
